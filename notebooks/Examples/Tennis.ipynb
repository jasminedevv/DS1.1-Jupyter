```python
import pandas as pd
from math import floor
df = pd.read_csv('../datasets/tennis.txt', 
                 delimiter="\t", 
                 header=None, 
                 names=['Outlook', 'Temp', 'Humidity', 'Wind', 'Decision'])
```

```python
df
```

```python
df[df['Wind'] == 'Weak']
df[df['Wind'] == 'Weak']['Decision']
def conditional_prob(c1, c2, condition):
    df_new = df[df[c1] == condition][c2]
    s = df_new.unique()
    population_size = len(df_new)
    pr = {}
    for i in s:
        pr[i] = len(df[(df[c1] == condition) & (df[c2]== i)]) / population_size

    print("The players are", floor(pr['Yes'] * 100), "% likely to play if the", c1, "is", condition)
```

```python
conditional_prob('Wind', 'Decision', 'Strong')
conditional_prob('Outlook', 'Decision', 'Rain')
conditional_prob('Temp', 'Decision', 'Cool')
conditional_prob('Humidity', 'Decision', 'High')
```
